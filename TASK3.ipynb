# **TASK 3: Visualisation Using Histogram**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('IRIS.csv')
data
```

```python
data.dtypes
```

```python
plt.hist(data['sepal_length'], bins=10, edgecolor='k', alpha=0.7)
plt.xlabel('Sepal Length Cm')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```
